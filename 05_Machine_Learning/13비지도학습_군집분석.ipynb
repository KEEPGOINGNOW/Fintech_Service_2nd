# 군집분석(계층적 군집, 비계층적 군집)
### 계층적 군집
  - 비지도학습 : 라벨 없이 유사 기반으로 그룹화
  - 데이터가 적을 때 유리
  - **거리(Metric)** 방식
      - 거리 계산 지표: 유크리드, 맨하탄, 코사인 유사도
  - 연결(Linkage) 방식 존재
  
### 계층적 군집 방법 5가지

  - **단일연결법**(single linkage/ 최단연결법)
      - 두 클러스터 간 최단 거리(가장 가까운 점/데이터) 기준으로 합병
      - 군집과 군집 사이에서 가장 가까운 데이터를 기준으로 묶음
      
  - **완전연결법**(complite linkage / 최장연결법)
      - 두 클러스터 간 최장 거리(가장 먼 점/데이터) 기준으로 합병
      - 군집과 군집 사이에서 가장 먼 데이터를 기준으로 묶음
      
  - **평균연결법**(average linkage)
      - 두 클러스터 간 모든 점/데이터 쌍의 평균 거리 기준으로 합병
      - 군집과 군집 사이의 모든 데이터의 평균 데이터로 계산(이상치에 덜 민감)
      
  - **중심연결법**(centroid linkage)
      - 각 클러스터 중심점 간 거리 기준으로 합병
      - 군집과 중심점 사이의 거리를 거리로 측정한 방식
      - 계산이 빠름
    
  - **와드연결법**(ward linkage)
      - 합변 후 클러스터 내 분산(오차 제곱합)이 최소가 되도록 합병
      - 군집내 오차가 최소가 되는 데이터로 계산
      - 군집내 분산이 최소이면서 조밀한 군집
      
#### **덴드로그램**으로 계층적 군집의 시각화 수행

```python
import scipy.cluster.hierarchy as shc
```

```python
# 원핫인코딩을 통한 object 데이터 타입 컬럼 변경
data = pd.get_dummies(data, drop_first=True)
data
```

```python
# 데이터 복사를 통한 원본 데이터셋 변화 최소화
data2 = data.copy()
```

```python
data = data.drop('CustomerID', axis=1)
data
```

### 단일연결법/최단연결법

```python
# 단일연결법으로 연결된 댄드로그램 트리 시각화
plt.figure(figsize=(20,10))
plt.title('Single linkage Dendrogram')
dend = shc.dendrogram(shc.linkage(data, method='single'))
plt.show()
```

### 완전연결법/최장연결법

```python
# 완전 연결법으로 연결된 댄드로그램 트리 시각화
plt.figure(figsize=(20,10))
plt.title('Complete linkage Dendrogram')
dend = shc.dendrogram(shc.linkage(data, method='complete'))
plt.show()
```

### 평균연결법

```python
# 평균 연결법으로 연결된 댄드로그램 트리 시각화
plt.figure(figsize=(20,10))
plt.title('Average linkage Dendrogram')
dend = shc.dendrogram(shc.linkage(data, method="average"))
plt.show()
```

### 중심연결법 centroid

```python
# 중심 연결법으로 연결된 댄드로그램 트리 시각화
plt.figure(figsize=(20,10))
plt.title('Centroid linkage Dendrogram')
dend = shc.dendrogram(shc.linkage(data, method="centroid"))
plt.show()
```

### 와드 연결법

```python
# 와드 연결법으로 연결된 댄드로그램 트리 시각화
plt.figure(figsize=(20,10))
plt.title('Ward linkage Dendrogram')
dend = shc.dendrogram(shc.linkage(data, method="ward"))
plt.show()
```

# 비계층적 군집 - 데이터 수가 많을 때 사용

### 비계층적 군집 분석
- 사전에 정한 군집 수(k)에 따라 데이터를 k개의 그룹으로 나눔

### 비계층적 군집 분석 방법
- k-means(k 평균 군집)
    - 각 군집 중심으로부터 거리(주로 유클리드)를 기반으로 군집을 형성
    - 이상치에 민감함
    - 데이터 로드 > 스케일링(StandardScaler를 사용 데이터에 이상치가 있으면 RobustScaler) > 
    - 초기화(군집 수 설정) > 할당(단일 연결법) > 업데이트(평균 위치로 중심점 이동) > 수렴 확인
    
- K-Medians
    - 중앙값을 이용해서 군집 형성
    - 이상치 완화
    - 데이터 로드 > 초기화(군집 수 설정) > 할당(단일 연결법) > 업데이트(중앙값으로 위치로 중심점 이동) > 수렴 확인

# 군집의 최적 개수 정하기
* elbow method
* 실루엣 지수

* 군집군석의 응집도(요소들이 얼마나 모여있는지 판단하는 지표)
  * inertia_ 값이 작을 수록 군집이 잘 형성됨
* elbow method: inertia_(응집도)를 출력하고 시각화한 뒤 응집도의 감소폭이 급격한 변화가 시작되는 지점을 최적 군집의 수로 채택하는 방법

```python
print(km.inertia_)
```

```python
inertia = []
for i in range(1, 21):
    km = KMeans(n_clusters=i, random_state=42)
    km.fit(ss_data)
    inertia.append(km.inertia_)
inertia
```

엘보우 메소드로 계산한 최적 군집 수: 3개

```python
plt.figure(figsize=(20,10))
sns.lineplot(x=range(1, 21), y=inertia, marker='o', markersize=10, markerfacecolor='red')
plt.show()
```

* 실루엣 지수를 출력해서 최적 군집 개수 구하기
* 실루엣 지수를 구하기 위해서는 최소 2개 이상의 군집이 필요
* 실루엣 점수가 가장 높은 클러스터 개수가 최적 군집 개수 (1에 가까울 수록 좋음)

```python
from sklearn.metrics import silhouette_score
```

```python
sil_scores = []
for i in range(2, 21):
    km2 = KMeans(n_clusters=i, random_state=42)
    km2.fit(ss_data)
    lables = km2.labels_
    sil_scores.append(silhouette_score(ss_data, lables))
sil_scores
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x=range(2, 21), y=sil_scores, marker='o', markersize=10, markerfacecolor='red')
plt.show()
```
