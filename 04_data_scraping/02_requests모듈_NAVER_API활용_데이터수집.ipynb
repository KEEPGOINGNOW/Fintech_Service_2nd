### 네이버 검색 API 활용
- 책을 검색하고 결과를 데이터프레임으로 생성
- 검색어는 input()을 이용해 입력받은 값을 이용

```python
import requests
import pandas as pd
```

```python
def json_unpack(data):
    result={}
    for item in data['items']:
        for key, value in item.items():
            result.setdefault(key, []).append(value)
    df = pd.DataFrame(result)
    return df
```

## for문 이용하여 전체 페이지 가져오기

```python
keyword = str(input('검색하고 싶은 키워드를 입력하세요: '))
```

```python
 all_data = []
```

```python
import requests

url = "https://openapi.naver.com/v1/search/book.json"
Client_Id = "_loifYh633NryTnmS7ZK"
Client_Secret = '7J5zAUYkd9'
headers = {'X-Naver-Client-Id': Client_Id, 'X-Naver-Client-Secret': Client_Secret}
payload = {'query': keyword, 'display': 100, 'start': 1, 'sort': 'sim'}
r = requests.get(url, params=payload, headers=headers)
print(r.url)
print(r.status_code)
data = r.json()

n_data = data['total']
total_pages = (int(data['total'])//100)+1

print('전체 검색 데이터 수: ', n_data, '검색할 페이지 수: ', total_pages)
result_df = json_unpack(data)
all_data.append(result_df)
```

```python
start = 101
for page in range(total_pages):
    print(f"{page+1}/{total_pages} 수집중", end='\r')
    url = "https://openapi.naver.com/v1/search/book.json"
    Client_Id = "_loifYh633NryTnmS7ZK"
    Client_Secret = '7J5zAUYkd9'
    
    payload = dict(query=keyword, display=100, start=start, sort="sim")
    headers = {'X-Naver-Client-Id': Client_Id, 'X-Naver-Client-Secret': Client_Secret}
    r = requests.get(url, params=payload, headers=headers)
    print(r.url)
    print(r.status_code)
    data = r.json()

    n_data = data['total']
    total_pages = (int(data['total'])//100)+1
#     print('전체 검색 데이터 수: ', data['total'], '검색할 페이지 수: ', (int(data['total'])//100+1))
    result_df = json_unpack(data)
    all_data.append(result_df)
    if start == 1000:
        break
    elif start < 901:
        start += 100
    elif start == 901:
        start += 99
```

```python
len(all_data)
```

```python
final_result = pd.concat(all_data)
final_result = final_result.reset_index(drop=True)
final_result
```
