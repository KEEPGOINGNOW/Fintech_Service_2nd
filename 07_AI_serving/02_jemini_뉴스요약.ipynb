```python

"""
🤖 Gemini AI 챗봇 데모 (Gradio Blocks)
- google-genai 클라이언트 사용
- Gradio Chatbot 컴포넌트 UI 예시
"""

import os
import gradio as gr
from google import genai

# API 키 로드
api_key = os.getenv("GEMINAI_API_KEY")
client = genai.Client(api_key=api_key)

# 채팅 응답 함수

def gemini_chat(user_message, chat_history):
    if not chat_history:
        messages = [{"role": "system", "content": "당신은 친절한 AI 어시스턴트입니다."}]
    else:
        messages = []
        for user, bot in chat_history:
            messages.append({"role": "user", "content": user})
            messages.append({"role": "assistant", "content": bot})
    messages.append({"role": "user", "content": user_message})

    response = client.models.generate_content(
        model="gemini-2.0-flash",
        contents=messages
    )
    bot_reply = response.text.strip()
    chat_history.append((user_message, bot_reply))
    return chat_history, chat_history

# Gradio 인터페이스 정의

def build_interface():
    with gr.Blocks() as demo:
        gr.Markdown("## 🤖 Gemini AI 챗봇 데모")
        chatbot = gr.Chatbot(label="Gemini AI")
        user_input = gr.Textbox(placeholder="메시지를 입력하세요...", show_label=False)
        send_btn = gr.Button("전송")

        send_btn.click(
            fn=gemini_chat,
            inputs=[user_input, chatbot],
            outputs=[chatbot, chatbot]
        )
        user_input.submit(
            fn=gemini_chat,
            inputs=[user_input, chatbot],
            outputs=[chatbot, chatbot]
        )
    return demo

# 데모 객체 생성 (실행 없이 인터페이스를 미리 봅니다)
demo = build_interface()
demo.launch()
```

```python
import os
import requests
import html
import pandas as pd
import gradio as gr
from google import genai

# 환경 설정
Client_Id = "_loifYh633NryTnmS7ZK"
Client_Secret = '7J5zAUYkd9'
api_key = os.getenv("GEMINAI_API_KEY")
client = genai.Client(api_key=api_key)

# 텍스트 정제 함수
def text_clean(x):
    x = html.unescape(x.replace('<b>', '').replace('</b>', ''))
    return x.strip()

# JSON -> DataFrame
def json_unpack(data):
    result = {}
    for item in data['items']:
        for key, value in item.items():
            result.setdefault(key, []).append(value)
    df = pd.DataFrame(result)
    return df

# 뉴스 수집 함수
def fetch_news(keyword):
    url = "https://openapi.naver.com/v1/search/news.json"
    headers = {'X-Naver-Client-Id': Client_Id, 'X-Naver-Client-Secret': Client_Secret}
    payload = {'query': keyword, 'display': 100, 'start': 1, 'sort': 'sim'}
    r = requests.get(url, params=payload, headers=headers)
    data = r.json()
    df = json_unpack(data)
    for col in ['title', 'description']:
        df[col] = df[col].apply(text_clean)
    return df[['title', 'originallink']]

# Gemini 요약 함수
def summarize_article(url):
    prompt = f"다음 뉴스 기사를 요약하고 소제목 포함 500자 내외로 정리해줘:\n{url}"
    response = client.models.generate_content(
        model="gemini-2.0-flash",
        contents=prompt  # ✅ 문자열만 전달
    )
    return response.text.strip()

# Gradio UI 정의
def build_interface():
    with gr.Blocks() as demo:
        gr.Markdown("## 📰 Gemini 뉴스 요약 데모")

        keyword_input = gr.Textbox(label="뉴스 키워드 입력", placeholder="예: 스테이블코인")
        search_btn = gr.Button("🔍 뉴스 검색")
        news_dropdown = gr.Dropdown(label="기사 제목 선택")
        summary_output = gr.Textbox(label="Gemini 요약 결과", lines=15)
        link_output = gr.Textbox(label="원문 URL")

        state_df = gr.State()

        def search_news(keyword):
            df = fetch_news(keyword)
            titles = df['title'].tolist()
            return gr.update(choices=titles), df

        def summarize_selected(title, df):
            selected_row = df[df['title'] == title].iloc[0]
            url = selected_row['originallink']
            summary = summarize_article(url)
            return summary, url

        search_btn.click(fn=search_news, inputs=keyword_input, outputs=[news_dropdown, state_df])
        news_dropdown.change(fn=summarize_selected, inputs=[news_dropdown, state_df], outputs=[summary_output, link_output])

    return demo

# 실행
if __name__ == '__main__':
    demo = build_interface()
    demo.launch()
```
