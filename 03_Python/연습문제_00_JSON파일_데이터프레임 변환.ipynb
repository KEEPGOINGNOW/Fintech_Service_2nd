```python
with  open ('./data/가상게임앱_리뷰데이터.json', 'r', encoding='utf-8') as file:
    review = file.read()
# print(data[:100])
```

```python
import json

data = json.loads(review)
data
```

```python
result = {}
for key in data[0]:
#     print(key)
    temp = []
    for dicts in data:
#         print(dicts[key])
        temp.append(dicts[key])
#     print(temp)
    result[key] = temp
# print(result)

import pandas as pd
result_df = pd.DataFrame(result)
len(result_df)
result_df
```

```python
result2 = {}
for land in data:
#     print(land)
    for key, value in land.items():
#         print(key, value)
        result2.setdefault(key, []).append(value)
# print(result2)

import pandas as pd
result2_df = pd.DataFrame(result2)
result2_df
```
