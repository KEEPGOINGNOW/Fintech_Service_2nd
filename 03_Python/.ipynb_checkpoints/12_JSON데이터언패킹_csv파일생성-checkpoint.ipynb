```python
with open('./data/com_kbstar_kbbank.json', 'r', encoding='utf-8') as file1:
    data = file1.read()
print(data[:100])
```

```python
#json 파일 변환 -> str
import json
reviews = json.loads(data)
print(len(reviews))
print(reviews[0])
```

```python
# 내 풀이
reviews_dict = {}
for key in reviews[0]:
#      print(key)
    review_list = []
    for dicts in reviews:
#         print(key, dicts[key])
        review_list.append(dicts[key])
#     print(review_list)
    reviews_dict[key] = review_list

import pandas as pd
review_df = pd.DataFrame(reviews_dict)
# review_df
```

```python
reviews_dict = {}
for key in reviews[0]:
#      print(key)
    review_list = []
    review_list2 = []
    for dicts in reviews:
#         print(key, dicts[key])
        if key == '별점':
            review_list.append(int(dicts[key]))
        else:
            temp = str(dicts[key]).replace('!','').replace(',','').replace('.','').replace('(','').replace(')','').replace('?','')
            review_list.append(temp)
#     print(review_list)
    reviews_dict[key] = review_list

import pandas as pd
review_df = pd.DataFrame(reviews_dict)
review_df
```

1. data 변수에 json.loads()로 문자열을 읽어온 파일 확인
2. 구조 파악하기 : 리스트-딕셔너리 구조로 구성
3. data는 리스트 안에 딕셔너리 세트로 저장됨
4. 각 딕셔너리에는 키로 json 파일의 정보가 저장
5. 딕셔너리에서 하나의 키당 딕셔너리의 갯수의 vlaue를 리스트 형태로 저장
6. pandas 이용해 dataframe으로 변환

```python
# 순서대로 진행해보기
# json 파일 변환한 것 확인
print(reviews[0])

# 각 리스트 생성
user = []
score = []
review = []
date = []
reponse = []

# 몇 개의 키로 구성되어 있는지 확인
len(reviews)
reviews[0].keys()
```

```python
# for문으로 len 수만큼 반복하여 각 리스트에 value 값을 넣음
for i in reviews:
#     print(count)
    user.append(i['사용자'])
    score.append(int(i['별점']))
    review.append(i['리뷰'])
    date.append(i['날짜'])
    reponse.append(i['개발자 응답'])

# print(user[0])
# print(score[0])
# print(review[0])
# print(date[0])
# print(reponse[0])
```

```python
review_dict = dict(사용자=user, 별점=score, 리뷰=review, 날짜=date, 개발자응답=reponse)
# review_dict

review_df = pd.DataFrame(review_dict)
review_df
```

```python
result2 = {}
for review in reviews:
#     print(review)
    for key, value in review.items():
#         print(key, value)
        result2.setdefault(key, []).append(value)
    
result2_df = pd.DataFrame(result2)
# result2_df
```

### 전국치킨집리스트.json


```python
with open('./data/전국치킨집리스트.json', 'r', encoding='utf-8') as file2:
    chicken = file2.read()
# print(chicken[:1000])
```

```python
cl = json.loads(chicken)
len(cl)
```

- cl에는 리스트 요소 37개 있음
- 리스트 요소 1개에는 1000개의 가게 정보가 있음
- 가게 정보는 cl[0]['body']['items']에 있음
- 가게 정보를 구성하는 key는 39개 있음

```python
# cl[0].keys()
# cl[0]['header'].keys()
cl[0]['body'].keys()
```

```python
%%time

result3 = {}
for page in cl:
    for item in page['body']['items']:
        for key, value in item.items():
    #         print(key, value)
            result3.setdefault(key, []).append(value)
    # print(len(result3['bizesId']))


result3_df = pd.DataFrame(result3)
result3_df
```

```python
cl[0]['header']['columns']
```

```python
%%time

# key값을 한글로 만든 key로 저장

result4 = {}
for page in cl:
    for item in page['body']['items']:
        for key, value in zip(page['header']['columns'],item.values()):
    #         print(key, value)
            result4.setdefault(key, []).append(value)
    # print(len(result3['bizesId']))


result4_df = pd.DataFrame(result4)
result4_df
```

```python
%%time

# 내 풀이
cl_dict = {}
for page in cl:
    for key in page:
    #     print(key) 
        if key == 'body':
            for key2 in page[key]:
    #             print(key2)
                if key2 == 'items':
                    for i in range(len(page[key][key2])):
                        for key3, value in page[key][key2][i].items():
                            if type(value) == int:
                                cl_dict.setdefault(key3, []).append(int(value))
                            else:
                                 cl_dict.setdefault(key3, []).append(value)

# import pandas as pd
cl_df = pd.DataFrame(cl_dict)
cl_df
```
