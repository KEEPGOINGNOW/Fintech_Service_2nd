# 리스트 표현식
- 리스트 안에 for 반복문과 if else 조건문을 한 반에 적어서 리스트를 만드는 방법
- 여러 줄의 코드를 1줄로 줄일 수 있다
- for문과 if else문을 사용해서 처리할 때 속도가 빠름
- [실행할 명령 for 변수 in list, tuple]

```python
# 정규표현식 사용
for i in range(1,11):
    print(f'{i}는 짝수입니다') if i % 2 == 0 else print(f'{i}는 홀수입니다')
```

```python
# 1부터 10까지 들어있는 리스트 생성
l1 = [] 

for i in range(1,11):
    l1.append(i)
print(l1)
```

```python
# 리스트 표현식
[i for i in range(1,11)]
```

## 리스트 표현식과 일반 반복문 속도 차이


```python
import time
```

```python
%%time
# start = time.time()
a = []
for i in range(1,20000000):
    a.append(i+5)
print(a[:10])

# print('코드 실행 시간: ', time.time() - start)
```

```python
%%time
# start2 = time.time()
b = [i+5 for i in range(20000000)]
print(b[:10])

# print('코드 실행 시간: ', time.time() - start)
```

- 1에서 10까지의 숫자 중 짝수만 넣어서서 리스트 생성


```python
l1 = []
for i in range(1,11):
    if i % 2 == 0:
        l1.append(i)
l1
```

## 다중 반복문을 사용한 리스트 표현식
- [i * j for i in ragne(2,10) for j in range(1,10)]

```python
gugudan = []
for i in range(2,10):
    for j in range(1,10):
        gugudan.append(f"{i} X {j} = {i * j}")
gugudan[:9]
```

```python
# 누적해서 사용하는 반복문에는 자주 사용하지 않음
# 그 이상은 잘 쓰지 않음
gugudan2 = [f"{i} X {j} = {i * j}" for i in range(2,10) for j in range(1,10)]  
gugudan2[9:18]
```

```python
# 시간
# 튜플로도 생성이 가능
time = [f"{i}:{j}:{k}" for i in range(0,23) for j in range(0,60) for k in range(0,60)]
time[:10]
```

# 파일 읽기 쓰기

### 1) open() : 파일에 문자열 쓰거나 읽는 경우
- 파일저장
    - 변수명 = open('파일이름', 모드(w:쓰기, r:읽기))
    - 변수명.write('저장할 내용')
    - 변수명.close() : close()를 안하면 계속 열려있음

```python
file = open('text.txt', 'w')
file.write('파일 만들기 연습')
file.close()
```

- 파일 열기
    - 변수명 = open('파일이름', 파일모드(r))
    - 변수2 = 변수1.read()
    - 변수1.close()
    

```python
file1 = open('text.txt', 'r')
my_file = file1.read()
file1.close()
print(my_file)
```

### 2)  with open()
- with open은 close()를 하지 않아도 자동으로 close() 처리

```python
with open('text2.txt', 'w') as file3:
    file3.write('with open을 사용하면 파일을 자동으로 닫아 편리함')
```

```python
with open('text2.txt', 'r') as file4:
    data = file4.read()
print(data)
```

# encoding에 따른 읽기 오류
- 운영체제나(windows, linux, macOS 등), 한국/영어 윈도우에 따라 인코딩 포멧 다름
- 인코딩에 따른 오류가 발생 가능
- web 표준은 utf-8이므로 인코딩은 utf-8로 하는 것이 좋음
- windows의 경우 인코딩이 cp949이므로 주의

```python
with open ('text_utf-8.txt', 'w', encoding='utf-8') as file5:
    file5.write('이것은 with open으로 만들 utf-8 인코딩 파일입니다.')
```

```python
with open ('text_utf-8.txt', 'r', encoding='utf-8') as file6:
    file6_data = file6.read()
print(file6_data)
```

# json 파일 읽어와서 내용 정리


```python
with open('./data/서울특별시_관광지입장정보_2011_2016.json', 'r', encoding='utf-8') as file7:
    trip_info = file7.read()
# print(trip_info)
print(type(trip_info))
```

```python
import json

data = json.loads(trip_info)
print(len(data))
data[:2]
```

```python
for item in data[:2]:
#     print(item)
    for key in item.values():
        print(key)
```

- value를 리스트로 생성 후 dict함수로 딕셔너리 생성하여 데이터프레임으로 변환

```python
year_mon = []
sido = []
place = []
foriners = []
total = []

for item in data:
    foriners.append(item['ForNum'])
    total.append(item['NatNum'])
    place.append(item['resNm'])
    sido.append(item['sido'])
    year_mon.append(item['yyyymm'])

    
tour_info = dict(년월=year_mon, 시도=sido, 장소=place, 외국인=foriners, 전체방문객수=total)
tour_info['년월'][:10]
```

```python
import pandas as pd

# 데이터프레임 할당
df_info = pd.DataFrame(tour_info)
df_info
```

- 딕셔너리를 바로 생성하여 데이터프레임으로 변환
- key, value를 직접 할당해서 딕셔너리 만들기

```python
result = {}
for key in data[0]:
    value_list=[]
#     print(key)
    for item in data:
        value_list.append(item[key])
#     print(key, value_list)
    result[key] = value_list
# print(result)
```

```python
result_df = pd.DataFrame(result)
result_df
```

- setdafault 함수를 이용해 생성
    - key가 있으면 새로운 값을 추가하면 

```python
dic_result = {}
for dicts in data:
    for key, value in dicts.items():
#         print(key, value)
        dic_result.setdefault(key,[]).append(value)
dic_result
```

```python
dic_result_df = pd.DataFrame(dic_result)
dic_result_df
```
